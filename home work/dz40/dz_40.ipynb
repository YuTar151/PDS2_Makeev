Проведіть класифікацію на основі датасету UCI ML Breast Cancer Wisconsin, використовуючи вказані у презентації моделі бустингу, беггінгу та стекінгу. Змініть їх параметри наступним чином:
### У stackingClassifier як базові алгоритми встановіть LogisticRegression та DecisionTreeClassifier, як фінальну модель - SVC;

```python
from sklearn.ensemble import StackingClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
data = load_breast_cancer()
```

```python
X, y = load_breast_cancer(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=12)
```

```python
estimators = [('lgr', LogisticRegression()), ('sgc', DecisionTreeClassifier())]
model_cls = StackingClassifier(estimators=estimators, final_estimator=SVC())
```

```python
model_cls.fit(X_train, y_train)
```

```python
print(model_cls.score(X_test, y_test))
```

### У BaggingClassifier як базовий алгоритм встановіть DecisionTreeClassifier;

```python
from sklearn.ensemble import BaggingClassifier
```

```python
model_bagcls = BaggingClassifier(base_estimator=DecisionTreeClassifier(), n_estimators=20, random_state=12)
model_bagcls.fit(X_train, y_train)
```

```python
model_bagcls.score(X_test, y_test)
```

### У AdaBoostClassifier як базовий алгоритм встановіть LogisticRegression


```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model_ada_bclf = AdaBoostClassifier(base_estimator=LogisticRegression(), n_estimators=100, random_state=12)
model_ada_bclf.fit(X_train, y_train)
```

```python
model_ada_bclf.score(X_test, y_test)
```

### У XGBClassifier як базовий алгоритм (booster) встановіть 'gbtree'.

```python
# !pip install xgboost
```

```python
from xgboost import XGBClassifier
```

```python
model_xgdb_clf = XGBClassifier(booster='gbtree', learning_rate=0.1, n_estimators=10)
model_xgdb_clf.fit(X_train, y_train)
```

```python
model_xgdb_clf.score(X_test, y_test)
```

### У зведеній таблиці порівняйте результати класифікацій, наведених у коді, із класифікаціями, що отримані вами.

```python
import pandas as pd
import numpy as np

models = ['StackingClassifier',
          'BaggingClassifier',
          'AdaBoostClassifier',
          'XGBClassifier']
score1 = [model_cls.score(X_test, y_test),
          model_bagcls.score(X_test, y_test),
          model_ada_bclf.score(X_test, y_test),
          model_xgdb_clf.score(X_test, y_test)]
les_scores = ["0.923976",
              "0.935672",
              "0.918128",
              "0.900584"]                

df = pd.DataFrame({'Models': models,
                   'HW Scores': score1,
                  'Lesson Scores': les_scores})

print(df)
```
