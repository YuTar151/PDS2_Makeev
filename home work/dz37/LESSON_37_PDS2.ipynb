#  Поліноміальна регресія в Python

```python
x = [2, 3, 4, 5, 6, 7, 7, 8, 9, 11, 12]
y = [18, 16, 15, 17, 20, 23, 25, 28, 31, 30, 29]
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
model = np.poly1d(np.polyfit(x, y, 3))
polyline = np.linspace(1, 12, 50)
plt.scatter(x, y)
plt.plot(polyline, model(polyline))
plt.show()
```

```python
print(model)
```

```python
from sklearn.metrics import mean_squared_error
x=10*np.random.normal(0,1,70)
y=10*(-x**2)+np.random.normal(-100,100,70)

plt.figure(figsize=(10,5))
plt.scatter(x,y,s=15)
plt.xlabel('Predictor',fontsize=16)
plt.ylabel('Target',fontsize=16)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x.reshape(-1,1), y.reshape(-1,1))
```

```python
y_pred = lm.predict(x.reshape(-1,1))
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x,y,s=15)
plt.plot(x,y_pred,color='r')
plt.xlabel('Predictor',fontsize=16)
plt.ylabel('Target',fontsize=16)
plt.show()
```

```python
print('RMSE for Linear Regression=>',np.sqrt(mean_squared_error(y,y_pred)))
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline

Input=[('polynomial',PolynomialFeatures(degree=2)),('modal',LinearRegression())]
pipe=Pipeline(Input)
pipe.fit(x.reshape(-1,1),y.reshape(-1,1))
```

```python
poly_pred=pipe.predict(x.reshape(-1,1))
sorted_zip = sorted(zip(x,poly_pred))
x_poly, poly_pred = zip(*sorted_zip)

plt.figure(figsize=(10,6))
plt.scatter(x,y,s=15)
plt.plot(x,y_pred,color='r',label='Linear Regression')
plt.plot(x_poly,poly_pred,color='g',label='Polynomial Regression')
plt.xlabel('Predictor',fontsize=16)
plt.ylabel('Target',fontsize=16)
plt.legend()
plt.show()
```

```python
print('RMSE for Polynomial Regression=>',np.sqrt(mean_squared_error(y,poly_pred)))
```

# Регулярізація

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from numpy import genfromtxt
dataset = genfromtxt('https://raw.githubusercontent.com/m-mehdi/tutorials/main/boston_housing.csv', delimiter=',')
X = dataset[:,:-1]
y = dataset[:,-1]
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0)
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
print(f"Score for train set {lr.score(X_train, y_train)}")
print(f"Score for test set {lr.score(X_test, y_test)}")
```

## L1

```python
from sklearn.linear_model import Lasso

lasso1 = Lasso(alpha=1.0).fit(X_train, y_train)
print(f"Score for train set {lasso1.score(X_train, y_train)}")
print(f"Score for test set {lasso1.score(X_test, y_test)}")
```

```python
lasso2 = Lasso(alpha=0.01).fit(X_train, y_train)
print(f"Score for train set {lasso2.score(X_train, y_train)}")
print(f"Score for test set {lasso2.score(X_test, y_test)}")
```

## L2

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=0.7).fit(X_train, y_train)
print(f"Score for train set {ridge.score(X_train, y_train)}")
print(f"Score for test set {ridge.score(X_test, y_test)}")
```

## L2 + L1

```python
from sklearn.linear_model import ElasticNet
eklastic = ElasticNet(alpha=0.01, l1_ratio=0.01).fit(X_train, y_train)
print(f"Score for train set {eklastic.score(X_train, y_train)}")
print(f"Score for test set {eklastic.score(X_test, y_test)}")
```
