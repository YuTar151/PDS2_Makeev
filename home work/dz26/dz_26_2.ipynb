```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Введемо значення спостережень в масив

```python
a = np.array([10, 13, 10, 9, 9, 12, 12, 6, 7, 9, 8, 9, 11, 9, 14, 13, 9, 8, 8, 7, 10, 10, 11, 11, 11, 12, 8, 7, 9, 10, 14, 13, 8, 8, 9, 10, 11, 11, 12, 12])
```

Відсортуємо значення:

```python
a.sort()
```

```python
b = np.unique(a, return_counts=True)
```

```python
data = pd.DataFrame({"variable": b[0], "frequency": b[1]})
```

побудуємо статистичний розподіл частот

```python
data["rel_freq"] = [i/data["frequency"].sum() for i in data["frequency"]]   #відносна частота
pr = data.to_string(index = False)
print(pr)
```

```python
data.plot(x="variable", y = "frequency", kind="line")
plt.grid()
```

Знайдемо моду і медіану вибірки:

```python
median = data["variable"].median()
print("Median is equal:", median)
```

```python
moda = float(data['variable'][data['frequency'] == data["frequency"].max()])
print('moda is equal', moda)
```
