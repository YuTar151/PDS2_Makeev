Використовуючи модель ARIMA і датасет Yearly sunspots data 1700-2008 побудуйте прогноз на 2009-2030 р.р.

Детальніше про даний датасет: 
https://www.statsmodels.org/dev/datasets/generated/sunspots.html

Імпорт датасету: 
statsmodels. api. datasets. sunspots. load_pandas ()

```python
import warnings
import itertools
import pandas as pd
import numpy as np
import statsmodels
import statsmodels.api as sm
import statsmodels.tsa.api as smt
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
data = statsmodels.api.datasets.sunspots.load_pandas()
ds = data.data
```

```python
ds
```

```python
ds.isnull().sum()
```

```python
plt.plot(ds['YEAR'], ds['SUNACTIVITY'])
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(ds['SUNACTIVITY'])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
import pmdarima as pm

# створення часового ряду з даними
ts = pd.Series(ds['SUNACTIVITY'].values, index=pd.to_datetime(ds['YEAR'], format='%Y'))

# автоматичний підбір параметрів моделі ARIMA
model = pm.auto_arima(ts, seasonal=False, suppress_warnings=True)

# визначення найкращих параметрів моделі ARIMA
results = sm.tsa.ARIMA(ts, order=model.order).fit()
print(results.summary())
```

```python
# прогнозування значень на період 2009-2030 рр.
forecast = results.forecast(steps=22)
```

```python
# візуалізація даних та прогнозу
plt.figure(figsize=(12,6))
plt.plot(ts, color='blue')
plt.plot(forecast, color='red')
plt.xlabel('YEAR')
plt.ylabel('SUNACTIVITY')
plt.title('Прогнозування активності сонця за допомогою моделі ARIMA')
plt.show()
```
