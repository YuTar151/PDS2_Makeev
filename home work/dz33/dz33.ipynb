### Створіть xls-документ, що містить дані про марки автомобілів, їхню вартість, споживання палива, максимальну швидкість та Вантажопідйомність. Кількість записів у документі – 20.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

car_brands = ['BMW', 'Audi', 'Mercedes-Benz', 'Toyota', 'Honda', 'Nissan', 'Ford', 'Chevrolet', 'Mazda', 'Subaru', 'Volvo', 'Volkswagen', 'Hyundai', 'Kia', 'Lexus', 'Cadillac', 'Jeep', 'Dodge', 'Ferrari', 'Tesla']
car_prices = [50000, 45000, 60000, 35000, 30000, 25000, 40000, 35000, 28000, 32000, 50000, 45000, 28000, 26000, 55000, 70000, 45000, 40000, 300000, 80000]
fuel_consumptions = [8.5, 9.0, 7.5, 6.0, 7.0, 8.0, 9.5, 10.0, 8.0, 8.5, 7.0, 6.5, 7.5, 7.0, 9.0, 12.0, 10.0, 11.0, 15.0, 4.0]
max_speeds = [250, 260, 270, 220, 220, 230, 240, 220, 220, 230, 240, 250, 200, 200, 250, 280, 240, 250, 320, 250]
load_capacities = [500, 450, 400, 600, 550, 500, 550, 600, 500, 450, 600, 550, 500, 500, 450, 400, 550, 600, 100, 500]

df = pd.DataFrame({
    'Brand': car_brands,
    'Price': car_prices,
    'Fuel Consumption': fuel_consumptions,
    'Max Speed': max_speeds,
    'Load Capacity': load_capacities
})

print(df)
```

### споживання палива від вартості - лінійною діаграмою;

```python
sns.lineplot(data=df, x='Price', y='Fuel Consumption')
```

### споживання палива від швидкості точковим графіком

```python
plt.plot(df['Fuel Consumption'], df['Max Speed'], 'o')
plt.title('Залежність споживання палива від швидкості')
plt.xlabel('Споживання палива, л/100 км')
plt.ylabel('Максимальна швидкість')
plt.show()
```

### Вантажопідйомності від швидкості точковим графіком

```python
df.head()
```

```python
sns.scatterplot(x='Load Capacity', y ='Max Speed', data = df)
```

### Побудуйте гістограму для розділення автомобілів на 5 груп за швидкістю

```python
plt.hist(df['Max Speed'], 5, width=8, color='g')
plt.xlabel('Максимальна швидкість')
plt.ylabel('Кількість автомобілів')
plt.title('Розподіл автомобілів за максимальною швидкістю')
plt.show()
```

### Побудуйте коробковий графік для відображення швидкостей різних марок

```python
sns.boxplot(x='Max Speed', y="Brand", data=df)
```
