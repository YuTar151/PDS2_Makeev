### Задача 1

Прилад містить 7 елементів, з яких три зношені. При включенні приладу випадковим чином включаються три елементи.
Знайти ймовірність того, що при цьому виявляться включеними два незношені елементи.


Розрахуємо скільки є всього способів включити три елементи з серед 7-и, при умові, що 2 з них не зношені.

Включаємо 2 незношені та 1 зношений: цей варіант може вийти у 7 * 4 * 3 = 84 способи.

Включаємо 3 незношені: цей варіант може вийти у 4 * 3 * 2 = 24 способи.
Всього ми маємо 7 * 6 * 5 = 210 способів включити три елементи. 

Тому імовірність того, що включаться два незношені елементи, складає:
(84 + 24) / 210 = 0.34285714285714286
Тобто близько 34%

### Задача 2

Для сигналізації про аварію встановлено два незалежно працюючих сигналізатори. Ймовірність того, що при аварії спрацює перший сигналізатор дорівнює 0,95, ця ймовірність для другого сигналізатора дорівнює 0,9. Знайти ймовірність того, що при аварії спрацює:

а) хоча б один сигналізатор;

б) тільки один сигналізатор.

а) Ймовірність того, що при аварії спрацює хоча б один сигналізатор:
P(A or B) = P(A) + P(B) - P(A * B)
P(A or B) = 0.95 + 0.9 - (0.95 * 0.9) = 0.995
Тобто імовірність того, що при аварії спрацює хоча б один сигналізатор, рівна 99.5%.

б) Подія A – "при аварії спрацює тільки один сигналізатор", це означає, що перший спрацює, а другий ні або другий спрацює, а перший ні.

Введемо на розгляд події

A1 – "спрацював перший" => P(А1) = 0,95

А2 – "спрацював другий" => P(А2) = 0,9

A1' – "перший не спрацював" => P(А1') = 1-0,95 = 0,05

A2' – "другий не спрацював" => P(А2') = 1-0,9 = 0,1

Оскільки сигналізатори працюють незалежно, то
P(A)= P(A1) * P(A2') + P(A1') * p(A2) = 0,95 * 0,1+0,05 * 0,9 = 0,095 + 0,045 = 0,14

При аварії спрацює тільки один сигналізатор з вірогідністю 14%
