Використовуючи датасет з файлу trayn.csv, побудуйте діаграми розподілу для значень з поля «Age».

Заповніть пропущені значення в датасеті з файлу trayn.csv. Обґрунтуйте вибір способу заповнення пропущених значень.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("trayn.csv")
data
```

```python
plt.hist(data["Age"], bins=50)
plt.title("Розподіл віку")
plt.xlabel("Вік")
plt.ylabel("Частота")
plt.show()
```

```python
count_empty = data["Age"].isnull().sum()
count_filled = data["Age"].count()
print("Кількість порожніх значень в стовпчику 'Age':", count_empty)
print("Кількість заповнених значень в стовпчику 'Age':", count_filled)
```

```python
median_age = data["Age"].median() 
data["Age"].fillna(median_age, inplace=True)
```

```python
count_empty = data["Age"].isnull().sum()
count_filled = data["Age"].count()
print("Кількість порожніх значень в стовпчику 'Age' після обробки:", count_empty)
print("Кількість заповнених значень в стовпчику 'Age' після обробки:", count_filled)
```

```python
data.to_csv("trayn_filled.csv", index=False)
```

Було обрано метод заповнення пропущених значень медіаною, оскільки він є простим і ефективним способом заповнення пропущених даних, що не впливає на форму розподілу даних. Метод заповнення пропущених значень медіаною є більш надійним та гнучким варіантом для заповнення порожніх значень віку.
