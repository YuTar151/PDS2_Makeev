
Із генеральної сукупності, розподіленої за нормальним законом, отримано вибірку об'ємом n = 20

|Xi| 1 | 2 | 4 | 6 | 7 |
|-:|--:|--:|--:|--:|--:|
|ni| 2 | 3 |10 | 3 | 2 |

Знайти з надійністю у = 0,95 інтервальну оцінку для середнього квадратичного відхилення $\sigma$ генеральної сукупності.


Для розрахунку інтервальної оцінки середнього квадратичного відхилення генеральної сукупності нам знадобиться розрахувати виборочне середнє значення:  
$$\overline X =  \frac{1}{N} \sum_{i=1}^{k} x_in_i$$   
А також виправлену дисперсію, яка розраховується за формулою:  
$$ D = \frac{\sum_{i=1}^{k} n_i(x_i - \overline X)^2}{N-1} $$   
де k - кількість варіант 5, N - об'єм вибірки, тобто 20.  
І стандартне відхилення:  $s = \sqrt{D}$,   
  
Щоб розрахувати довірчий інтервал для стандартного відхилення генеральної сукупності з надійністю 95% скористуємось формулою:  $$s(1-q) < \sigma < s(1+q)$$    
Де q отримуємо з таблиці спираючись на значення у = 0,95 та N = 20, значення q = 0,37


```python
import numpy as np
x = np.array([1, 2, 4, 6, 7])
n = np.array([2, 3, 10, 3, 2])
N = sum(n)
t = 1.96
q = 0.37

# Виборочне середнє
X = sum(x*n)/N 

# Дисперсія
D = sum([n[i]*(x[i] - X)**2 for i in range(len(x))])/(N-1)

# Стандартне відхилення
sd = round(D**0.5, 4)

# Довірчий інтервал
sd_range = [round(sd*(1 - q), 2), round(sd*(1 + q), 2)] 

print(f'Виборочне середнє значення дорівнює: {X}')
print(f"Дисперсія дорівнює: {round(D, 2)}")
print(f'Середнє квадратичне відхилення дорівнює: {sd}') 
print(f'Довірчий інтервал для стандартного відхилення: {sd_range}')
```

**Висновок:** Стандартне відхилення генеральної сукупності з ймовірністю 95% належить інтервалу [1.12 : 2.43].
